```python
import pandas as pd
import geopandas as gpd
import numpy as np

#LINHAS = gpd.read_file('Linhas2.geojson')

CON_EOL = pd.read_parquet('dataframeEOL2.parquet')
EOL = gpd.read_file('EOL.geojson')


UFV = gpd.read_file('UFV.geojson')
UFV['conjunto'] = UFV['conjunto'].fillna('none')
UFV.nome = UFV.nome.str.lower()
UFV.conjunto = UFV.conjunto.str.lower()



CON_UFV = pd.read_parquet('dataframeFV2.parquet')

CON_UFV['Corte %'].fillna(0, inplace=True)

CON_UFV.nom_usina = CON_UFV.nom_usina.str.lower()
CON_UFV['conjunto'] = CON_UFV.nom_usina
CON_UFV.drop(columns=['nom_usina'], inplace=True)
CON_UFV.conjunto = CON_UFV.conjunto.str.replace('fotovoltaico ', '', regex=False)

nomes = {
'conj. fotov. acu iii 230kv': 'conj. açú iii 230 kv',
'conj. três marias 3 138kv':'conj. três marias 3 138 kv',
'conjunto santa luzia ii 500kv':'conjunto santa luzia ii 500 kv',
}
CON_UFV.conjunto.replace(nomes, inplace=True)

#UFVnone = UFV[UFV['conjunto'] == 'none']
#UFVnone[UFVnone.tem_corte == True]

#get the names of CON_UFV.conjunto.unique() that aint in UFV.conjunto.unique()
#missing = set(CON_UFV.conjunto.unique()) - set(UFV.conjunto.unique())
#missing,len(missing)
```

```python
#drop duplicates in UFV.conjunto
UFV = UFV.drop_duplicates(subset=['conjunto'], keep='first')
mapa_subestacao = UFV.set_index('conjunto')['subestacao']
CON_UFV['subestacao'] = CON_UFV['conjunto'].map(mapa_subestacao)
```

```python
CON_UFV['Corte %'].groupby(CON_UFV['subestacao']).mean().sort_values(ascending=False).to_csv('constr_ufv.csv')
```
